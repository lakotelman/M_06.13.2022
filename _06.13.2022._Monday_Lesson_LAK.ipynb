Another thing I wanted to try was parsing it out. What if I accecpted one argument, a string containing the operation? I then could use the if statements and perform the calculations using the indexes of the list. 

**Couple issues:**

What if they put floats in? The amount of conditionals would get a little crazy to include both using this formula. 


```python
def calculator(equation): 
    e = equation.split(" ")
    if e[1] == "+": 
        return(int(e[0]) + int(e[2]))
    elif e[1] == "-":
        return(int(e[0]) - int(e[2]))
    elif e[1] == "*": 
        return(int(e[0]) * int(e[2]))
    elif e[1] == "/": 
        return(int(e[0]) / int(e[2]))
    else: 
        return "Try again. (Make sure to put spaces between your numbers and the operator.)"
        
calculator("4 - 3")
```

# Exercise 2 <br>
<p>Create a pyramid of X's for n number of levels.</p>

```python
#Simple triangle or a pyramid of x's 

#----X----  = 9 
#---XXX---  = 9
#--XXXXX--  = 9
#-XXXXXXX-  = 9
#XXXXXXXXX  = 9 

#hint: Pay attention to X's and Spaces
#For each iteration the spaces decrease by 1 from top to bottom 
#For each iteration the x's increase by 2 from top to bottom 
```

For this one I wanted the argument to take however many levels. Initially I had a -= for the i in the for loop and that didn't really work. Eventually I figured out I could use the range to step down.

```python
def pyramid(levels): 
    padding = levels-2
    space = " "
    exes = 'X'
    for i in range(1,levels+2,2): 
        print(f"{space*padding}{i*exes}")
        padding -= 1
        
pyramid(60)
```
